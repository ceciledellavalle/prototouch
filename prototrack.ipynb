```python
from pynput.mouse import Listener as MouseListener
from pynput.keyboard import Listener as KeyboardListener, Key

# Variable pour contrôler l'exécution
stop_execution = False

# Fonction qui sera appelée à chaque mouvement de la souris
def on_move(x, y):
    print(f"Mouvement de la souris: ({x}, {y})")
    if stop_execution:
        return False  # Arrêter le listener de la souris

# Fonction qui sera appelée à chaque clic de souris
def on_click(x, y, button, pressed):
    if pressed:
        print(f"Clic à ({x}, {y}) avec le bouton {button} pressé.")
    else:
        print(f"Clic à ({x}, {y}) avec le bouton {button} relâché.")
    if stop_execution:
        return False  # Arrêter le listener de la souris

# Fonction pour arrêter l'écoute du clavier et de la souris (quand la touche Escape est pressée)
def on_press(key):
    global stop_execution
    if key == Key.esc:  # Si la touche Escape est pressée
        print("Arrêt du programme...")
        stop_execution = True  # Mettre fin à l'exécution des listeners
        return False  # Cela arrête le listener du clavier aussi

# Lancer un Listener pour surveiller les événements de la souris
mouse_listener = MouseListener(on_move=on_move, on_click=on_click)

# Lancer un Listener pour surveiller les événements du clavier
keyboard_listener = KeyboardListener(on_press=on_press)

# Démarrer les deux listeners dans des threads séparés
mouse_listener.start()
keyboard_listener.start()

# Attendre que les deux listeners finissent
mouse_listener.join()  # Attendre la fin du listener de la souris
keyboard_listener.join()  # Attendre la fin du listener du clavier

print("Le programme a terminé.")
```
